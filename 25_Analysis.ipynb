## 找出最火的文章

```python
df_lite = df_all.filter(items=['title','publish_date','views','stars','likes','author_name','url'])
```

```python
df_lite.sort_values(by='views',ascending=False).head(10)
```

```python
df_lite.sort_values(by='stars',ascending=False).head(10)
```

```python
df_lite.sort_values(by='likes',ascending=False).head(10)
```

```python
author_df = df.groupby('author_name').sum()

author_df['count'] = df.groupby('author_name').count()._id

author_df['views_per_artical'] = author_df['views']/author_df['count']

author_df = author_df.sort_values(by='views_per_artical',ascending=False)

author_df.count()

author_df
```
