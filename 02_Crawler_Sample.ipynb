# 批量爬取网页
> 批量爬bilibili视频，并存储和分析

```python
# export
%reload_ext autoreload
%autoreload 2
import pandas as pd
import requests,json,re
from bs4 import BeautifulSoup,Tag,NavigableString
from collections import Counter
from crawler_from_scratch.utils import *
```

## 批量爬数据

```python
data = {}

for page in range(1,51):
    url = f'https://search.bilibili.com/all?keyword=%E5%85%AC%E5%BC%80%E8%AF%BE&from_source=banner_search&page={page}'
    print(url)
    res = requests.get(url,headers={'user-agent':'Mozilla/5.0'})
    if res.status_code == 200:
        soup = BeautifulSoup(res.text)
    else:
        print(res,res.text)
            
    main_content = soup.body.find(class_='video-list')
    
    for c in get_children(main_content):
        item_data = get_data(c)
        a_title_url = item_data['a_title_url']
        id = re.search(r'av\d+',a_title_url).group()
        data[id] = item_data
    
```

## 数据存取

```python
with open('./data/02_bilibili.json', 'w') as f:
    json.dump(data,f)
```

```python
with open('./data/02_bilibili.json', 'r') as f:
    data = json.loads(f.read())
```
